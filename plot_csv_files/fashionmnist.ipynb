```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to normalize power values
def normalize_power(value):
    if 'mW' in value:
        return float(value.replace('mW', ''))
    elif 'W' in value:
        return float(value.replace('W', '')) * 1000
    else:
        return float(value)

# Read the CSV file into a DataFrame
df = pd.read_csv('f3.csv')

# Remove '%' sign from the 'CPU' column
df['CPU'] = df['CPU'].str.replace('%', '').astype(float)

# Normalize the 'Power' column
df['Power'] = df['Power'].apply(normalize_power)

# Plotting all graphs in a single multiplot
plt.figure(figsize=(12, 8))

# CPU vs Time
plt.subplot(2, 2, 1)
plt.plot(df['Time'], df['CPU'])
plt.title('CPU vs Time')
plt.xlabel('Time')
plt.ylabel('CPU')

# Memory vs Time
plt.subplot(2, 2, 2)
plt.plot(df['Time'], df['Memory'])
plt.title('Memory vs Time')
plt.xlabel('Time')
plt.ylabel('Memory')

# Temperature vs Time
plt.subplot(2, 2, 3)
plt.plot(df['Time'], df['Temperature'])
plt.title('Temperature vs Time')
plt.xlabel('Time')
plt.ylabel('Temperature')

# Power vs Time
plt.subplot(2, 2, 4)
plt.plot(df['Time'], df['Power'])
plt.title('Power vs Time')
plt.xlabel('Time')
plt.ylabel('Power')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
